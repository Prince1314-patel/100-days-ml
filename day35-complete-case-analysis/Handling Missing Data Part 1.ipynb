For doing CCA(Complete Case Ananlysis) which means to remove whole rows which have missing value,the missing values should be missing at random places i.e MCAR(Missing Completely At Random)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_science_job.csv')
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
df.shape
```

```python
# extractiong rows which have missing values
cols = [var for var in df.columns if df[var].isnull().mean() < 0.05 and df[var].isnull().mean() > 0]
cols
```

```python
df[cols].sample(5)
```

```python
# analysing perecntage of data lost
(len(df[cols].dropna()) / len(df)) * 100
```

```python
new_df = df[cols].dropna()
df.shape, new_df.shape
```

```python
new_df.hist(bins=50, density=True, figsize=(12,12))
plt.show()
```

```python
for cols in new_df.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    # original data
    df[cols].hist(bins=50, density=True, color='red')
    plt.title(cols)

    # data after cca, the argument alpha makes the color transparent, so we can
    # see the overlay of the 2 distributions
    new_df[cols].hist(bins=50, color='green', density=True, alpha=0.8)  
```

```python

temp = pd.concat([
            # percentage of observations per category, original data
            df['enrolled_university'].value_counts() / len(df),

            # percentage of observations per category, cca data
            new_df['enrolled_university'].value_counts() / len(new_df)
        ],
        axis=1)

# add column names
temp.columns = ['original', 'cca']

temp
```

```python
temp = pd.concat([
            # percentage of observations per category, original data
            df['education_level'].value_counts() / len(df),

            # percentage of observations per category, cca data
            new_df['education_level'].value_counts() / len(new_df)
        ],
        axis=1)

# add column names
temp.columns = ['original', 'cca']

temp
```
