```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("student_clustering.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=df['cgpa'],y=df['iq'])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
X = df.iloc[:,:].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)
```

```python
y_means
```

```python
plt.scatter(x=X[y_means == 0,0],y=X[y_means == 0,1],color="blue")
plt.scatter(x=X[y_means == 1,0],y=X[y_means == 1,1],color="green")
plt.scatter(x=X[y_means == 2,0],y=X[y_means == 2,1],color="red")
plt.scatter(x=X[y_means == 3,0],y=X[y_means == 3,1],color="black")
```
