```python
df = df.drop("Date",axis=1)
```

```python
df.head()
```

```python
#let's have a look at daily confirmed cases 
import plotly.express as px
fig = px.bar(df,x="Date_YMD", y="Daily Confirmed")
fig.show()
```

```python
#now let's visualise death rate due to covid

cases = df['Daily Confirmed'].sum()
deceased = df['Daily Deceased'].sum()

labels=["Confirmed", "Deceased"]
values = [cases,deceased]

px.pie(df,values=values,names=labels,title="Daily confirmed cases vc daily deaths",hole=0.5)
```

```python
death_rate = (df["Daily Deceased"].sum() / df["Daily Confirmed"].sum()) * 100
print(death_rate)
```

```python
import plotly.express as px
fig = px.bar(df, x='Date_YMD', y='Daily Deceased')
fig.show()
```

the AutoTS library, which is one of the best Automatic Machine Learning libraries for Time Series Analysis

```python
#Now here’s how to predict covid-19 deaths with machine learning for the next 30 days:
#!pip install autots

from autots import AutoTS
model = AutoTS(forecast_length=30, frequency='infer', ensemble='simple')
model = model.fit(df, date_col="Date_YMD", value_col='Daily Deceased', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
