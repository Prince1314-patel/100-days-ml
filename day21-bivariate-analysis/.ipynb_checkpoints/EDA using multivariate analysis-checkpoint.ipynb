```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
titanic = pd.read_csv('train.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

# 1.Scatterplot(Numerical-Numerical)

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.scatterplot(x=tips ['total_bill'],y=tips['tip'],hue=tips['smoker'])
```

# 2. bar Plot(Numerical-Categorical)

```python
titanic.head()
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Fare'],hue=titanic['Sex'])
```

# 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

# 4. Distplot (Numerical - Categorical)

```python
sns.distplot(x=titanic[titanic['Survived']==0]['Age'],hist=False,color='red')
sns.distplot(x=titanic[titanic['Survived']==1]['Age'],hist=False)
```

# 5. HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

# 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

# 7. Pairplot

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

# 8. Lineplot (Numerical - Numerical)

```python
flights.head()
```

```python
sns.lineplot(x=flights['year'],y=flights['passengers'],hue=flights['month'])
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
